## Data Cleaning

### Missing values for Age

```python
training_processed['Age'].hist()
```

```python
#Use median for each passenger class - generate lookup table for age medians
#Create array of zeros
#2x3 matrix - 2 genders, 3 passenger classes
median_age_class_gender = np.zeros((2,3))

for i in range(0,2):
    if i==0:
        gender = 'male'
    else:
        gender = 'female'
    for j in range(0,3):
        median_age_class_gender[i,j] = training_processed[(training_processed.Pclass == j+1) & (training_processed.Sex == gender)]['Age'].dropna().median()

#Initalise Age fill column with age
training_processed['Age_fill'] = training_processed['Age']

for i in range(0,2):
    if i==0:
        gender = 'male'
    else:
        gender = 'female'
    for j in range(0,3):
        training_processed.loc[(training_processed.Age.isnull()) & (training_processed.Sex == gender) & (training_processed.Pclass == j+1),'Age_fill'] = median_age_class_gender[i,j]
        
training_processed[training_processed.Age.isnull()].head(10)
```

## Feature Engineering 

### Gender

* male = 0
* female = 1

```python
training_processed['Gender'] = training_processed['Sex'].map({'male':0, 'female':1}).astype(int)
training_processed.head(10)
```

### Title

```python
#Generate Title field
#Aim: from name field want to extract the title field. 
#Split on comma and extract title
training_processed['Title'] = training_processed['Name'].str.split(',').str[1].str.split('.').str[0]

training_processed['Title'].value_counts()
    
```

### Family size

```python
training_processed['Family_size'] = training_processed['Parch'] + training_processed['SibSp']
training_processed.head(10)
```
